```python
from bangla_stemmer.stemmer import stemmer
stmr = stemmer.BanglaStemmer()
def stem_bangali(message):
    stemmedArr = [stmr.stem(term) for term in message.split(" ")]
    return ' '.join(stemmedArr)
```

```python
df["text"] = df["text"].apply(stem_bangali)
```

```python
X_train,X_test,y_train,y_test = train_test_split(df["text"],df["label"],test_size=0.2,random_state=42)
```

```python
count_vect = CountVectorizer()
X_train_count = count_vect.fit_transform(X_train)
```

```python
X_test_count = count_vect.transform(X_test)
```

```python
X_train_count.shape
```

```python
X_test_count.shape
```

```python
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_count)
X_train_tfidf.shape
```

```python
X_test_tfidf = tfidf_transformer.transform(X_test_count)
X_test_tfidf.shape
```

```python
reg_clf = LogisticRegression(max_iter=1000).fit(X_train_tfidf, y_train)
```

```python
predicted = reg_clf.predict(X_test_tfidf)
np.mean(predicted == y_test)
```

```python
svc_clf = SVC(kernel='linear', probability=True).fit(X_train_tfidf, y_train)
```

```python
predicted = svc_clf.predict(X_test_tfidf)
np.mean(predicted == y_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB().fit(X_train_tfidf, y_train)
predicted = clf.predict(X_test_tfidf)
np.mean(predicted == y_test)
```
